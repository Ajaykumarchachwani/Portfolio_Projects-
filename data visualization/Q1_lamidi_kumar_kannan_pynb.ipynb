# **UMAP to visualize dataset **

```
#Performing UMAP to visualize dataset 
theumap= umap.UMAP(n_neighbors= 5, min_dist=0.1, n_components= 2, metric='euclidean')
the1umap= theumap.fit_transform(OUP_scld)
figures = list(signmn['label'])

data1 = [gr.Scatter(x= the1umap[:,0], y= the1umap[:,1], mode= 'markers',
                   marker= dict(color=OUP, colorscale='Rainbow', opacity=0.5),
                   text=[f'digit: {a}' for a in figures],
                   hoverinfo= 'text')]
                   
lay = gr.Layout(title= 'UMAP DIMENSIONALITY-REDUCTION', width = 1000, height= 1000,
                   xaxis = dict(title= 'Umap_1'),
                   yaxis = dict(title= 'Umap_2'))
fig1 = gr.Figure(data=data1,layout=lay)
off.plot(fig1, filename='UMAP8.html')
```
